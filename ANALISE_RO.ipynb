```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 5)

df_mar = pd.read_csv('SINASC_RO_2019_MAR.csv')
df_abr = pd.read_csv('SINASC_RO_2019_ABR.csv')
df_mai = pd.read_csv('SINASC_RO_2019_MAI.csv')
df_dez = pd.read_csv('SINASC_RO_2019_DEZ.csv')
```

```python
print(df_mar.columns)
print(df_abr.columns)
print(df_mai.columns)
print(df_dez.columns)
```

```python
for df in [df_mar, df_abr, df_mai, df_dez]:
    df['DTNASC'] = pd.to_datetime(df['DTNASC'])

def grafico_nascimentos_por_dia(df, mes):
    df['DTNASC'].dt.day.value_counts().sort_index().plot(kind='bar', color='skyblue')
    plt.title(f'Nascimentos por dia - {mes}')
    plt.xlabel('Dia do ms')
    plt.ylabel('Quantidade de nascimentos')
    plt.show()

def grafico_peso(df, mes):
    sns.histplot(df['PESO'], kde=True, bins=30, color='lightgreen')
    plt.title(f'Distribuio do peso ao nascer - {mes}')
    plt.xlabel('Peso (g)')
    plt.ylabel('Frequncia')
    plt.show()

grafico_nascimentos_por_dia(df_mar, "Maro")
grafico_peso(df_mar, "Maro")

grafico_nascimentos_por_dia(df_abr, "Abril")
grafico_peso(df_abr, "Abril")

grafico_nascimentos_por_dia(df_mai, "Maio")
grafico_peso(df_mai, "Maio")

grafico_nascimentos_por_dia(df_dez, "Dezembro")
grafico_peso(df_dez, "Dezembro")
```
